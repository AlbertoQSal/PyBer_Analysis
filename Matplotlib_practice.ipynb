# Create a Line Chart using the Matlab Method


```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

## Annotate Charts Using the MATLAB Method

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set the x limit between jan and june
# plt.xlim(0, 5) or plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

# Create a Line Chart using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

## Annotate Chart Using the Object-Oriented Interface

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label = "Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

# Create a Bar Chart Using the MATLAB Approach

## Create a vertical chart

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.show()
```

## Create a Horizontal Bar Chart

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# To invert the axis so the months could be in order we use the gca() method
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label="Boston")
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.legend()
plt.show()
```

# Create Bar Charts Using the Object-Oriented Approach

## Create a Vertical Bar Chart

```python
# Create the bar chart with ax.plt()
#fig, ax = plt.subplots()
#ax.bar(x_axis, y_axis)

# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the bar chart with ax.plt()
ax = plt.axes()
ax.bar(x_axis, y_axis, color="green", label = "Boston")
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.legend()
```

## Create a Horizontal Bar Chart

```python
# Create the bar chart with ax.plt()
#fig, ax = plt.subplots()
#ax.bar(x_axis, y_axis)

# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the bar chart with ax.plt()
ax = plt.axes()
ax.barh(x_axis, y_axis, color="cyan", label = "Boston")
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.legend()
```

# Create Scatter Plots and Bubble Charts

## Create a Bubble Chart Using the MATLAB Method

```python
# Scatter plots using the plt.plot method
plt.plot(x_axis, y_axis, 'o')
```

```python
# Scatter plots using the plt.scatter method
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.gca().invert_yaxis()
plt.xlim(0,45)
plt.legend()
plt.show()
```

# Create a Bubble Chart Using the MATLAB Method 2.0

```python
#Increase the size of th bubble using a for cycle
# MATLAB Method
y_axis_larger = []

for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

```python
# We can also use the comprehension list: new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
plt.show()
```

# Creat a Scatter Plot Using the Object-Oriented Interface

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis], color = 'skyblue', 
           alpha=0.8, edgecolor='k', label='Boston')
ax.invert_yaxis()
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
```

# Creat a Pie Chart Using the MATLAB Method

```python
# Matlab
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8)) #Increases the size of the figure
plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct='%.1f%%', colors = colors)
plt.show()
```

# Create a Pie Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,explode=explode_values, colors=colors, autopct='%.1f%%')
plt.show()
```
