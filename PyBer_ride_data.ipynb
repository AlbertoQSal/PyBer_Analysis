```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df['Month'])
plt.show()
```

```python
# To create a bar chart with the same data there are two ways: 
# First one:
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Second one: 
# Standard deviation
stdev = statistics.stdev(pyber_ride_df['Avg. Fare ($USD)'])
# Plot
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color = 'skyblue',
                   yerr=stdev, kind='bar', capsize=3, label='Avg. Fare ($USD)')

plt.show()
```
